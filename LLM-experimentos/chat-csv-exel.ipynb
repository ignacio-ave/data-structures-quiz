```python
from typing import TextIO
from langchain.agents import create_csv_agent, create_pandas_dataframe_agent
from langchain.llms import OpenAI
```

```python


def excel_to_csv():
    """
    Pide el archivo Excel y el archivo CSV al usuario, y convierte el archivo Excel a CSV.
    """
    print("Convertir archivo Excel a CSV")

    excel_file = input("Por favor, introduce la ruta del archivo Excel que deseas convertir: ")
    print(f"Archivo Excel seleccionado: {excel_file}")

    csv_file = input("Por favor, introduce la ruta donde deseas guardar el archivo CSV: ")
    print(f"Ruta de archivo CSV seleccionada: {csv_file}")

    data_xls = pd.read_excel(excel_file, index_col=None)
    data_xls.to_csv(csv_file, encoding='utf-8', index=False)

    print(f"Archivo Excel convertido a CSV exitosamente y guardado en {csv_file}")


def get_answer_csv():
    """
    Pide el archivo CSV y la consulta al usuario, y devuelve la respuesta a la consulta del archivo CSV.
    """
    
    
    csv_file = input("Por favor, introduce la ruta del archivo CSV que deseas consultar: ")
    print(f"Ruta del archivo CSV: {csv_file}")
    
    df = pd.read_csv(csv_file)
    
    # Crear el agente
    
    # agent = create_csv_agent(OpenAI(model_name="gpt-3.5-turbo", temperature=0), csv_file, verbose=False)
    agent = create_pandas_dataframe_agent(OpenAI(model_name="gpt-3.5-turbo", temperature=0.009), df, verbose=True)

    while True:
        query = input("Por favor, introduce tu consulta: " + ". Responde en español")
        print(f"Consulta: {query}")


        # Correr la consulta
        answer = agent.run(query)
        # verbose=False en la definicion del agente
        #print("Espere un momento por favor...")
        #time.sleep(25)
        
        print(f"\nRespuesta a la consulta: {answer}")
        
        another_query = input("¿Deseas hacer otra consulta a este archivo CSV? (S/N): ")
        if another_query.lower() != 's':
            break


```

```python
def main():
    # Mapa de opciones del menú
    menu_options = {
        '1': excel_to_csv,
        '2': get_answer_csv,
        '3': exit
    }

    while True:
        print("\nMenú:")
        print("1. Convertir archivo Excel a CSV")
        print("2. Realizar una consulta a un archivo CSV")
        print("3. Salir")
        option = input("Por favor, selecciona una opción: ")

        time.sleep(2)
        
        # Agarrar la función correspondiente a la opción
        function = menu_options.get(option)

        # Cheaquear si la opción es válida
        if function:
            # Ejecutar la función correspondiente
            # Si la función es get_answer_csv, imprimir la respuesta
            if function is get_answer_csv:
                print(f"Respuesta a la consulta: {function()}")
            elif function is exit:
                break
            else:
                function()
        else:
            print("Opción inválida. Por favor, vuelve a intentarlo.")

if __name__ == "__main__":
    main()
```
