```python
def imprimir_socrates_gpt(tema, objetivo, con_codigo, formato):
    print(f"\n\n")
    print(f"Nombre: Socrates-gpt")
    print(f"Funcion: Generar preguntas socraticas consecutivas una por una para ir midiendo y ayudando a entender en mayor profundidad los temas tratados.")
    print(f"Objetivos: {objetivo}")
    print(f"Instrucciones adicionales: ['debes llevar el conteo de las respuestas correctas consecutivas', 'si la respuesta es erronea, argumenta el porque de una manera que facilite la comprecion de los conceptos erroneos','debes llevar el conteo de las respuestas erroneas consecutivas', 'si el usuario lleva mas de tres respuestas consecutivas erroneas se debe mostrar una cheatsheet del tema y la materia tratada y bajar el nivel de dificultad de las preguntas drasticamente']")
    print(f"El tema tratado en las preguntas es: {tema}")
    print(f"Las preguntas deben estar pensadas para realizar este objetivo: {objetivo}")
    print(f"Intrucciones adicionales : ['debes llevar el conteo de las respuestas correctas consecutivas', 'si la respuesta es erronea, argumenta el porque de una manera que facilite la comprecion de los conceptos erroneos','debes llevar el conteo de las respuestas erroneas consecutivas', 'si el usuario lleva mas de tres respuestas consecutivas erroneas se debe mostrar una cheatsheet del tema y la materia tratada y bajar el nivel de dificultad de las preguntas drasticamente' ]")
    
    # print(f"Las preguntas deben incluir código" if con_codigo else "")
    free(con_codigo)
    free(formato)
    # print(f"Las preguntas deben ser de este tipo: ['preguntas abiertas', 'preguntas cerradas', 'preguntas de opcion multiple']" if formato else "" )
    
    print(f"Debes presentarte como socrates-gpt y personificar a socrates.")
    
def elegir_tema():
    print("Elija un tema de la siguiente lista: ")
    

    for i in range(len(estructuras_de_datos)):
        print(f"{i+1}. {estructuras_de_datos.iloc[i]['Estructura de datos']}")

        
        
    
    time.sleep(1)
    
    tema = input("Ingrese el tema que desea estudiar: ")
    tema = int(tema)
    tema = estructuras_de_datos.iloc[tema-1]['Estructura de datos']
    
    return tema

def elegir_objetivo():
    
    print("Que busca lograr en esta sesion de estudio?")
    time.sleep(1)
    objetivo = input("Ingrese el objetivo que desea lograr: ")
    return objetivo

def with_codigo():
    print("Desea que las preguntas incluyan codigo?")
    time.sleep(1)
    codigo = input("Ingrese si o no: ")
    if codigo == "si" or codigo == "Si" or codigo == "SI":
        return True
    else:
        return False
    
def distinto_formato():
    print("Desea que las preguntas sean de distinto formato?")
    time.sleep(1)
    formato = input("Ingrese si o no: ")
    if formato == "si" or formato == "Si" or formato == "SI":
        return True
    else:
        return False
```

```python
# main  
def main():
    temas = elegir_tema()
    objetivo = elegir_objetivo()
    codigo = with_codigo()
    formato = distinto_formato()
    
    print("\n\n")
    
    imprimir_socrates_gpt(temas, objetivo, codigo, formato)

    if temas in estructuras_datos_operaciones:
        print(f"Las operaciones asociadas con {temas} son:")
        for operacion in estructuras_datos_operaciones[tema]:
            print(f"- {operacion}")    
```
